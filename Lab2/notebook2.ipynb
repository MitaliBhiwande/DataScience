```r
install.packages("XML")
```

```r
library(XML)
```

```r
xmldata <- xmlParse("http://aiweb.cs.washington.edu/research/projects/xmltk/xmldata/data/courses/reed.xml")
```

```r
print(xmldata)
```

```r
xmlist <- xmlToList(xmldata)
```

```r
xml.df <- xmlToDataFrame(xmldata)
```

```r
# split the misaligned columns into separate data frames
place.df <- as.data.frame(t(xpathSApply(xmldata,"//place",function(x) xmlSApply(x,xmlValue))))
time.df<-as.data.frame(t(xpathSApply(xmldata,"//time",function(x) xmlSApply(x,xmlValue))))
```

```r
#delete the separated columns from original data frame
xml.df <- subset( xml.df, select = -c(place, time) )
```

```r
head(xml.df)
```

```r
# combine all the data frames to obtain the final data frame
newxml.df<-cbind.data.frame(xml.df,time.df,place.df)
```

```r
print(newxml.df)
```

```r
head(newxml.df)
```

```r
write.csv(file="xmltocsv.csv",newxml.df)
```
