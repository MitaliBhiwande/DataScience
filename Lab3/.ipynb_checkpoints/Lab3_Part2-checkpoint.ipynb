# Classical data: Edgar Andersonâ€™s Iris data

### Get iris data

```r
head(iris)
```

```r
names(iris)
```

### Assign data from column 1-4 and from column 5 to separate variables.

```r
x = iris[,-5]
y = iris$Species
```

### Create 3 clusters as per number of classes in iris data 

```r
kc <- kmeans(x,3)
```

```r
kc
```

```r
head(kc)
```

### Use 'table' to compare clustering result with the species/classes iris data to know the number of errors and missing data

```r
table(y,kc$cluster)
```

### Plot obtained clusters

```r
plot(x[c("Sepal.Length", "Sepal.Width")], col=kc$cluster)
points(kc$centers[,c("Sepal.Length", "Sepal.Width")], col=1:3, pch=23, cex=3)
```
