### Extract the hashtags and user mentions

```r
library(stringr)
```

```r
soccerdata= str_extract_all(soccertweets$text, "#\\w+")
soccerdata1= str_extract_all(soccertweets$text, "@\\w+")
dat<-soccertweets$text
dat2 <- unlist(dat)
dat3 <- grep("dat2", iconv(dat2, "latin1", "ASCII", sub="dat2"))
dat4 <- dat2[-dat3]
soccertext <- gsub(" ?(f|ht)tp(s?)://(.*)[.][a-z]+", "", dat4)
soccertext<- gsub("^ ", "", soccertext)
soccertext<- gsub(" $", "", soccertext)
#print(soccertext)
write(soccertext,"soccertext.txt")
```

```r
install.packages("SnowballC")
library(tm)
library(SnowballC)
library(wordcloud)
require(RColorBrewer)
```

```r
soccerdata <- read.csv('abc1.csv', stringsAsFactors = FALSE,comment.char="",fileEncoding="UTF-8")
```

```r
soccerdata[order(-soccerdata$V2),]
```

```r
pal2 <- brewer.pal(8,"Dark2")
wordcloud(word=soccerdata$V1, freq=soccerdata$V2, max.words =800,min.freq=5,scale=c(8,.5), 
           random.order = FALSE,rot.per=.5,vfont=c("sans serif","plain"),colors=pal2)
```
